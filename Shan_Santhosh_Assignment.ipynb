```python
#Since there are some circles above the upper limit there might be some outliers.
#Checking Outliers

Q1=np.percentile(data['Age'],25,interpolation='midpoint')
Q2=np.percentile(data['Age'],50,interpolation='midpoint')
Q3=np.percentile(data['Age'],75,interpolation='midpoint')
print('Q1=',Q1)
print('Q2=',Q2)
print('Q3=',Q3)
```

```python
#Calculating IQR
IQR= Q3 - Q1
print('IQR=',IQR)
```

```python
#Calculating upper and lower limit
low= Q1 - 1.5*IQR
up= Q3 + 1.5*IQR
print(low)
print(up)
```

```python
outlier=[]
for x in data['Age']:
    if((x>up) or (x<low)):
        outlier.append(x)
```
